### 🔍 Exploratory Analysis & Visualization
 Before diving into insights, it’s key to explore respondent demographics — like country, age, gender, education, and job status — to check how well the survey represents the global programming community. Large surveys often have some bias, so this step helps spot it early.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

analysis country wise 

```python
survey_df.Country.nunique() 
# counting total number of counties participated in total
```

```python
schema_df.Country
```

```python
top_countries_contibuted=survey_df['Country'].value_counts().head(10)
top_countries_contibuted
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=45)
plt.title("Top Contributing Countries")

ax = sns.barplot(x=top_countries_contibuted.index, y=top_countries_contibuted, palette="Set2")
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=3)

plt.tight_layout()
plt.show()
```

A large share of responses come from the USA and India, likely due to their large populations and widespread use of English among professionals. Since the survey is in English, it may underrepresent developers from non-English-speaking countries. Additionally, some regions may have lower awareness of the survey or rely less on platforms like Stack Overflow, further skewing the results.

 ### Gender
 Let's look at the distribution of responses for the Gender. It's a well known fact that women and non-binary genders are underrepresented in the programming community, so we might expect to see a skewed distribution here.

```python
schema_df.Gender
```

```python
gender_counts=survey_df.Gender.value_counts(dropna=False)
```

Only about 8% of survey respondents who have answered the question identify as women or non-binary. This number is lower than the overall percentage of women & non-binary genders in the programming community - which is estimated to be around 12%.

In which countries do developers work the highest number of hours per week? Consider countries with more than 250 responses only.

```python
survey_df.groupby('Country')[["WorkWeekHrs","Age"]].mean()
countries_df = survey_df.groupby('Country') [['WorkWeekHrs']].mean().sort_values('WorkWeekHrs',ascending=False)
high_response_countries_df = countries_df.loc[survey_df.Country.value_counts()>250].head(15)
high_response_countries_df
```

countries like Iran, China & Israel have the highest working hours, followed by the United States. However, there isn't too much variation overall and the average working hours seem to be around 40 hours per week.
